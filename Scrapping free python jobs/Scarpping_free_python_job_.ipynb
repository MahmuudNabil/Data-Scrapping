```python
#import libraries 
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#get page 
url="https://pythonjobs.github.io/"
page = requests.get(url)

#parse page using Beautiful soup
soup = BeautifulSoup(page.content, 'html.parser')

#job_lists 
job_lists = soup.find_all('div' , class_='job')
print(job_lists)
```

```python
#get data for all jobs
for job in job_lists:
  jobs_title = job.find('h1')
  jobs_info = job.find_all('span' , class_='info')
  jobs_location = jobs_info[0]
  jobs_date = jobs_info[1]
  jobs_type = jobs_info[2]
  jobs_company= jobs_info[3]
  jobs_desc = job.find('p' , class_ ='detail')
  jobs_link = jobs_title.find('a')
  
  print('job title  : ', jobs_title.text.strip())
  print('company location : ' , jobs_location.text.strip()  )
  print('Date : ', jobs_date.text.strip())
  print('type : ', jobs_type.text.strip())
  print('company  : ' , jobs_company.text.strip())
  print('job Description  : ' , jobs_desc.text.strip())
  print('job link  : ' , jobs_link['href'])
  print()

```

```python
#get only python jobs
python_jobs = soup.find_all('h1' , string= lambda text : 'python' in text.lower())
python_jobs

## access to parent of these jobs 
python_job_elements = [h1_element.parent for h1_element in python_jobs]
python_job_elements
```

```python
job_title = []
job_location =[]
date_post =[]
job_type =[]
company_name =[]
job_description = []
job_link =[]
```

```python
python_job_elements
```

```python
#get data for all jobs
for job_element in python_job_elements:
  title = job_element.find('h1')
  job_title.append(title.text.strip())

  jobs_info = job_element.find_all('span' , class_='info')
  location = jobs_info[0]
  job_location.append(location.text.strip())

  p_date = jobs_info[1]
  date_post.append(p_date.text.strip())

  jtype = jobs_info[2]
  job_type.append(jtype.text.strip())

  company= jobs_info[3]
  company_name.append(company.text.strip())

  desc = job.find('p' , class_ ='detail')
  job_description.append(desc.text.strip())

  link = jobs_title.find('a')
  job_link.append(link['href'])

  print('job title  : ', title.text.strip())
  print('company location : ' , location.text.strip()  )
  print('Date : ', p_date.text.strip())
  print('type : ', jtype.text.strip())
  print('company  : ' , company.text.strip())
  print('job Description  : ' , desc.text.strip())
  print('job link  : ' , link['href'])
  print()
```

```python
#save data into dataframe 
data = pd.DataFrame({"job_Title" :job_title ,
                     "Location" : job_location , 
                     "post_date" : date_post ,
                     "Job_type" : job_type , 
                     "Company" : company_name , 
                     "Description" : job_description ,
                     "link" : job_link})
data
```

```python
#save data into csv file formate
data.to_csv('real_python_jobs.csv')
```
