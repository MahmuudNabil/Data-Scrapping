```python
#import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#GET page 
url =  "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
page = requests.get(url)

#parse page 
soup = BeautifulSoup(page.content , 'html.parser')
soup
```

```python
#get table data 
my_table=soup.find('table', class_='wikitable sortable plainrowheaders')
print(my_table)
```

```python
#get states data 
table_header  = my_table.find_all('th')
names = []
for elem in table_header:
  link_tag = elem.find_all('a')       # get text from link 
  for txt in link_tag:
    names.append(txt.text)
print(names)
```

```python
# remove unwanted index 
  # index 9 and found ['D'] 
final_list = names[9:]
states =[]
for stri in final_list :
  if len(stri) > 3:           # chech for remove ['D']
    states.append(stri)
print(states)
```

```python
#get population Data 
divs = my_table.find_all('div')
pop = []
for txt in divs:
  pop.append(txt.string)
print(pop)
```

```python
#remove un wanted index like '7'
final_pop =[]
for i in pop:
  if len(i) > 3:
    final_pop.append(i)
print(final_pop)
```

```python
# store all data into Dataframe
data = pd.DataFrame({'State' : states,
                     'population' : final_pop})
data
```

```python
#save data into csv file
data.to_csv('countries_pop.csv')
```
