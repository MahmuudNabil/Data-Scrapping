```python
# import libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd 
import datetime
import time
```

```python
# Connect to Website and pull its data
def extract(page_num):
  headers = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
  url =f'https://wuzzuf.net/search/jobs/?a=navbl&q=BI%20Developer&start={page_num}'
  
  page = requests.get(url , headers)

  soup = BeautifulSoup(page.content , 'html.parser')
  
  return soup
```

```python
joblist = []
```

```python
def transform(soup):
  divs = soup.find_all('div' , {'class' : 'css-1gatmva e1v1l3u10'})
  
  for item in divs:
    title = item.find('a' , {'class' : 'css-o171kl'}).text.strip()
    company_name = item.find('a' , {'class' : 'css-17s97q8'}).text.strip()
    location = item.find('span' , {'class' : 'css-5wys0k'}).text.strip()
    post_date = item.find('div' , {'class' : 'css-d7j1kk'}).div.text.strip()
    job_type = item.find('div' , {'class' : 'css-1lh32fc'}).span.text.strip()
    
    skills_tag = item.find('div' ,{'class' : 'css-1lh32fc'}).next_sibling

    skills_list =[]
    for tag in skills_tag.find_all(['a' ,'span']):
      skills = tag.text.strip()
      skills_list.append(skills)
    
    job ={'title' : title ,
          'company_name' : company_name , 
          'post_date' : post_date,
          'job_type' : job_type,
          'skills' : skills_list 
          }

    joblist.append(job)
    
    print(joblist)
  return
```

```python
# looping on all pages
for page_num in range(0, 7,1):
  print(f'page number:  {page_num}')
  page = extract(page_num)
  transform(page)
```
